### 2.2 Creating dataframe
We use **pandas.DataFrame** to create dataframe<br>
**1. Syntax** : pandas.DataFrame(data, index=None, columns=None, dtype=None, Copy=None)<br>
**2. data** : Elements/Data to be stored in dataframe<br>
**3. index=** (Optional) : Index to use for output frame<br>
**4. columns=** : Labels for columns<br>
**5. dtype=** (Optional) : Provide the type of data<br>
**6. copy=** (Optional) : To copy input data<br>

```python
#using dictionary
data = {'names':['sanhci','abc','xyz'],'Branch':["IT","CSE","ECE"]}
dataframe = pd.DataFrame(data,columns=['names','Branch'])
dataframe
```

```python
#using Tuples
t=[("sanchi","it"),("abc","CSE"),("xyz","it")]
df=pd.DataFrame(t,columns=['name','branch'],index=['a','b','c'])
df
```
